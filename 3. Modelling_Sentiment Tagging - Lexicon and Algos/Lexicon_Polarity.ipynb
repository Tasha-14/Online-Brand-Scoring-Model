```python
import pandas as pd
import numpy as np
```

```python
#Input File
inputfile1 = 'Input/Data_Manual_Polarity.xlsx'
inputfile2 = 'Input/Keywords_With_Polarity.csv'
```

```python
df = pd.read_excel(inputfile1)
```

```python
df.shape
```

```python
df1=pd.read_csv(inputfile2)
df1.drop(['POS'], inplace=True, axis=1)
```

```python
Key_dict={}
```

```python
for i in range(len(df1['Lemma'])):
    Key_dict.update({df1['Lemma'][i]:df1['Polarity_Score'][i]})
```

```python
def split(sent):
    x=sent.split()
    return x
```

```python
df['Split']=df['Review'].apply(split)
```

```python
df['Matching_Keywords']=None
df['Polarity_Per_Word']=None
```

```python
sum=0
a=0    #a is sent tracker 
for i in df['Split']:
    b=0 #word tracker
    w=[]
    x=[]    
    for j in i:  
        limit=0
        for k in df1['Lemma']:
            if j==k and limit==0:
                #sum=sum+1
                #print(j,a,b)
                w.append(k)
                df['Matching_Keywords'][a]=w 
                x.append(Key_dict[k])                
                df['Polarity_Per_Word'][a]=x
                limit=1
                
        b=b+1
    a=a+1    
#print(sum,'\n')
```

```python
df['Polarity_Per_Word'].isnull().sum()
```

```python
df2 = df[df['Polarity_Per_Word'].notna()]
```

```python
df2
```

```python
def sumof(text):
    sum=0
    for i in text:
        sum=sum+i
    return sum      
```

```python
df2['Polarity_Per_Word'].apply(sumof)
```

```python
df2['Polarity_Sum_Lexicon'] = df2['Polarity_Per_Word'].apply(sumof)
```

Polarity through positive and negative words

```python
def count(val):
    count=0
    for i in val:
        if i>=0:
            count= count + 1
    return count      
```

```python
df2['Positive_Words']=df2['Polarity_Per_Word'].apply(count)
```

```python
def count1(val):
    count=0
    for i in val:
        if i<0:
            count= count + 1
    return count    
```

```python
df2['Negative_Words']=df2['Polarity_Per_Word'].apply(count1)
```

```python
df2['Pos-Neg'] = df2['Positive_Words']-df2['Negative_Words']
```

```python
pol=[]
for val in df2['Pos-Neg']:
    if val>=1:
        pol.append(1)
    elif val<1:
        pol.append(-1)
    else:
        pol.append(0)
df2['Word_Frequency_Polarity']=pol            
```

```python
df=df2
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
df['compound'] = [analyzer.polarity_scores(x)['compound'] for x in df['Review']]
df['neg'] = [analyzer.polarity_scores(x)['neg'] for x in df['Review']]
df['neu'] = [analyzer.polarity_scores(x)['neu'] for x in df['Review']]
df['pos'] = [analyzer.polarity_scores(x)['pos'] for x in df['Review']]
```

```python
score=[]
for value in df['compound']:
    if value<0:
        score.append(-1)
    elif value>0:
        score.append(1)
    else:
        score.append(0)
df['NLTK_Polarity'] = score       

df.drop(['compound','pos','neg','neu'],axis=1,inplace=True)
```
