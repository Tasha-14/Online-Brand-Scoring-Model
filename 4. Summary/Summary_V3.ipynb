```python
import pandas as pd
import numpy as np
```

```python
#Input File
summary_input = 'Input/Summary_Input.xlsx'
```

```python
df = pd.read_excel(summary_input)
df = df[['Sno', 'Source', 'Brand', 'Campaign', 'Review', 'Dimension', 'Theme','Topic', 'Manual_Polarity']]
```

```python
df
```

```python
brand = list(('Hector', 'Gloster', 'ZSEV', 'Hector Plus', 'Tata Harrier','Hyundai Creta'))
```

### Dimensions_Freq and Polarity

```python
res=[]
for i in brand:
    df1 = pd.DataFrame(df[df['Brand']==i].groupby([ 'Source','Campaign',"Dimension"] ).size().unstack().reset_index())
#df3= pd.DataFrame(df.groupby( [ "Brand", 'Source' ,'Campaign',"Topic"] ).size().unstack().reset_index())

    dd_pos = pd.DataFrame(df[df['Brand']==i].groupby([ 'Source','Campaign','Dimension'])['Manual_Polarity'].agg(lambda x: (x==1).sum()).unstack().reset_index())
    dd_pos = dd_pos.add_suffix('_Pos') 
    dd_neg = pd.DataFrame(df[df['Brand']==i].groupby([ 'Source','Campaign','Dimension'])['Manual_Polarity'].agg(lambda x: (x==-1).sum()).unstack().reset_index())
    dd_neg = dd_neg.add_suffix('_Neg') 
    dd_neu = pd.DataFrame(df[df['Brand']==i].groupby([ 'Source','Campaign','Dimension'])['Manual_Polarity'].agg(lambda x: (x==0).sum()).unstack().reset_index())
    dd_neu = dd_neu.add_suffix('_Neu') 
    dg=pd.concat([df1,dd_pos,dd_neg,dd_neu] ,axis=1)
    dg=dg.replace(to_replace=np.nan, value=0)
    dg['Brand']= i
    res.append(dg)
```

```python
res1= pd.DataFrame(res[0])
res2= pd.DataFrame(res[1])
res3= pd.DataFrame(res[2])
res4= pd.DataFrame(res[3])
res5= pd.DataFrame(res[4])
res6= pd.DataFrame(res[5])
```

```python
res1=res1[['Brand','Source', 'Campaign', 'Customer Service','Customer Service_Pos','Customer Service_Neg','Customer Service_Neu',
           'Design','Design_Pos', 'Design_Neg','Design_Neu','Interiors','Interiors_Pos','Interiors_Neg','Interiors_Neu',
       'Performance','Performance_Pos','Performance_Neg','Performance_Neu', 'Price',  'Price_Pos','Price_Neg',  
        'Price_Neu']]
res2=res2[['Brand','Source', 'Campaign', 'Customer Service','Customer Service_Pos','Customer Service_Neg','Customer Service_Neu',
           'Design','Design_Pos', 'Design_Neg','Design_Neu','Interiors','Interiors_Pos','Interiors_Neg','Interiors_Neu',
       'Performance','Performance_Pos','Performance_Neg','Performance_Neu', 'Price',  'Price_Pos','Price_Neg',  
        'Price_Neu']]
res3=res3[['Brand','Source', 'Campaign', 'Customer Service','Customer Service_Pos','Customer Service_Neg','Customer Service_Neu',
           'Design','Design_Pos', 'Design_Neg','Design_Neu','Interiors','Interiors_Pos','Interiors_Neg','Interiors_Neu',
       'Performance','Performance_Pos','Performance_Neg','Performance_Neu', 'Price',  'Price_Pos','Price_Neg',  
        'Price_Neu']]
res4=res4[['Brand','Source', 'Campaign', 'Customer Service','Customer Service_Pos','Customer Service_Neg','Customer Service_Neu',
           'Interiors','Interiors_Pos','Interiors_Neg','Interiors_Neu',
       'Performance','Performance_Pos','Performance_Neg','Performance_Neu', 'Price',  'Price_Pos','Price_Neg',  
        'Price_Neu']]
res5=res5[['Brand','Source', 'Campaign', 'Customer Service','Customer Service_Pos','Customer Service_Neg','Customer Service_Neu',
           'Design','Design_Pos', 'Design_Neg','Design_Neu','Interiors','Interiors_Pos','Interiors_Neg','Interiors_Neu',
       'Performance','Performance_Pos','Performance_Neg','Performance_Neu', 'Price',  'Price_Pos','Price_Neg',  
        'Price_Neu']]
res6=res6[['Brand','Source', 'Campaign', 'Customer Service','Customer Service_Pos','Customer Service_Neg','Customer Service_Neu',
           'Design','Design_Pos', 'Design_Neg','Design_Neu','Interiors','Interiors_Pos','Interiors_Neg','Interiors_Neu',
       'Performance','Performance_Pos','Performance_Neg','Performance_Neu', 'Price',  'Price_Pos','Price_Neg',  
        'Price_Neu']]
```

```python
dimension_freq = pd.concat([res1,res2,res3,res4,res5,res6]).reset_index()
del dimension_freq['index']
```
