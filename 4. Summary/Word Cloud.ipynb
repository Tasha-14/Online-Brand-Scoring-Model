```python
import pandas as pd
import numpy as np
```

```python
from wordcloud import WordCloud, STOPWORDS 
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
#Input File
summary_input = 'Input/Summary_Input.xlsx'
```

```python
df = pd.read_excel(summary_input)
df = df[['Sno', 'Source', 'Brand', 'Campaign', 'Review', 'Dimension', 'Theme','Topic', 'Manual_Polarity']]
```

```python
neg = df[df['Manual_Polarity']==-1].reset_index()
```

```python
del neg['index']
del neg['Sno']
```

```python
text = neg['Review'] 

wordcloud = WordCloud().generate(str(text))
```

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
